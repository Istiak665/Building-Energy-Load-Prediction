### 4. Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load scores
hl_scores = pd.read_csv('../results/tables/heating_load_scores.csv')
cl_scores = pd.read_csv('../results/tables/cooling_load_scores.csv')
```

```python
# Visualization for actual vs predicted values (Example for heating load)
# Assuming the predictions are stored in hl_scores and cl_scores
plt.figure(figsize=(10, 6))
plt.plot(hl_scores['fit_time'], label='Actual')
plt.plot(hl_scores['score_time'], label='Predicted')
plt.title('Actual vs Predicted Heating Load')
plt.xlabel('Fold')
plt.ylabel('Heating Load')
plt.legend()
plt.savefig('../results/figures/actual_vs_predicted_hl.png')
plt.show()
```

```python
# Visualization for relationship between metrics (Example for R2 and MAE for heating load)
plt.figure(figsize=(10, 6))
plt.scatter(hl_scores['test_R2'], hl_scores['test_MAE'])
plt.title('Relationship between R2 and MAE for Heating Load')
plt.xlabel('R2')
plt.ylabel('MAE')
plt.savefig('../results/figures/r2_vs_mae_hl.png')
plt.show()
```
